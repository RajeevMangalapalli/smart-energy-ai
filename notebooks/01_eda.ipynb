```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\rajee\Desktop\VS code\smart-energy-ai\data\processed\DE_load_daily.csv", index_col = "timestamp", parse_dates = ["timestamp"])
print(df.index.dtype)
```

```python
def plot_daily_load(df: pd.DataFrame):
    df["load_mw"].plot(figsize=(18, 6))
    plt.title("Daily Load in Germany")
    plt.ylabel("Load (MW)")
    plt.show()


plot_daily_load(df)
```

```python
# 30 day rolling average
df["load_mw_rolling"] = df["load_mw"].rolling(window=30).mean()
df[["load_mw", "load_mw_rolling"]].plot(figsize=(18, 6))
plt.title("Daily Load in Germany with 30-day Rolling Average")
plt.ylabel("Load (MW)")
plt.show()
```

```python
# Monthly average
df["monthly_average"] = df["load_mw"].resample("ME").transform("mean")
df[["load_mw", "monthly_average"]].plot(figsize=(18, 6))
plt.title("Daily Load in Germany with Monthly Average")
plt.ylabel("Load (MW)")
plt.show()
```
